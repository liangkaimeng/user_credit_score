# 导入模块

```python
import warnings
import pandas as pd
import seaborn as sns
from toad import quality
from toad.metrics import PSI
from toad.detector import detect
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")
pd.set_option("display.width", 10000)
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
```

# 加载数据

```python
train = pd.read_csv("../data/cs-training.csv", encoding="utf-8")
train.rename(columns={"Unnamed: 0": "Id"}, inplace=True)
test = pd.read_csv("../data/cs-test.csv", encoding="utf-8")
test.rename(columns={"Unnamed: 0": "Id"}, inplace=True)
```

# 替换字段

```python
map_fields = {"x1": "RevolvingUtilizationOfUnsecuredLines", 
              "x2": "age", 
              "x3": "NumberOfTime30-59DaysPastDueNotWorse", 
              "x4": "DebtRatio", 
              "x5": "MonthlyIncome", 
              "x6": "NumberOfOpenCreditLinesAndLoans", 
              "x7": "NumberOfTimes90DaysLate", 
              "x8": "NumberRealEstateLoansOrLines", 
              "x9": "NumberOfTime60-89DaysPastDueNotWorse", 
              "x10": "NumberOfDependents"}
train.rename(columns=dict(zip(map_fields.values(), map_fields.keys())), inplace=True)
test.rename(columns=dict(zip(map_fields.values(), map_fields.keys())), inplace=True)
```

## 数据分布

### 分布可视化

```python
i = 1
plt.figure(figsize=(40, 40), dpi=400)
for col in train.iloc[:, 2:].columns.tolist():
    ax = plt.subplot(4, 3, i)
    ax = sns.kdeplot(train[col], color="Red", shade=True)
    ax = sns.kdeplot(test[col], color="Blue", shade=True)
    ax.set_xlabel(map_fields[col])
    ax = ax.legend(["train", "test"])
    i += 1
    
plt.savefig(r"D:\liangkaimeng\picture\user_credit_score\credit_score_kaggle_distplot.png", bbox_inches='tight', dpi=300)
```

### PSI分布检验

```python
psi_df = PSI(train, test).reset_index()
psi_df["index"] = psi_df["index"].map(map_fields)
psi_df
```
